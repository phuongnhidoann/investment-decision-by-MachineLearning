```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import data
data = pd.read_excel('Data.xlsx', sheet_name='BCTC')
```

```python
#Calculates financial ratios
data['ROE'] = data['18. Lợi nhuận sau thuế (đồng) (Y)'] / data['D. VỐN CHỦ SỞ HỮU (đồng) (Y)']
data['ROA'] = data['18. Lợi nhuận sau thuế (đồng) (Y)'] / data['TỔNG TÀI SẢN (đồng) (Y)']
data['DFL'] = data['C. NỢ PHẢI TRẢ (đồng) (Y)'] / data['I. Nợ ngắn hạn (đồng) (Y)']
data['Current_Ratio'] = data['A. TÀI SẢN NGẮN HẠN (đồng) (Y)'] / data['I. Nợ ngắn hạn (đồng) (Y)']
data['D/E'] = data['C. NỢ PHẢI TRẢ (đồng) (Y)'] / data['D. VỐN CHỦ SỞ HỮU (đồng) (Y)']

data['EPS'] = data['18. Lợi nhuận sau thuế (đồng) (Y)'] / ((data['- Cổ phiếu phổ thông (đồng) (Y)']/10000) - (data['5. Cổ phiếu quỹ (đồng) (Y)']/10000))
data['P/E'] = 10000 / data['EPS']
data['BVPS'] = (data['D. VỐN CHỦ SỞ HỮU (đồng) (Y)'] - data['3. Tài sản cố định vô hình (đồng) (Y)']) / ((data['- Cổ phiếu phổ thông (đồng) (Y)']/10000) - (data['5. Cổ phiếu quỹ (đồng) (Y)']/10000))
data['P/B'] = 10000 / data['BVPS']
```

```python
#Create DataFrame contains calculated columns and 'Mã CK' and 'Năm' columns
df = data.loc[:, ['Mã CK', 'Năm', 'ROE', 'ROA', 'DFL', 'Current_Ratio', 'D/E',
                 'EPS', 'P/E', 'BVPS', 'P/B']]

df
```

```python
null_counts = df.isnull().sum()
null_counts
```

```python
df.fillna(0, inplace=True)
df.info()
```

```python
df = df[(df[['ROE', 'ROA', 'DFL', 'Current_Ratio', 'D/E',
             'EPS', 'P/E', 'BVPS', 'P/B']] != 0).all(axis=1)]

df['non_zero_year'] = df.groupby('Mã CK')['Năm'].transform(lambda x: (x != 0).any())
df = df[df['non_zero_year']]
df.drop('non_zero_year', axis=1, inplace=True)

df.info()
```

```python
#Columns to process
columns_to_process = ['ROE', 'ROA', 'DFL', 'Current_Ratio', 'D/E',
                      'EPS', 'P/E', 'BVPS', 'P/B']

#Calculate the average value of each column only from non-zero values
symbol_nonzero_mean_values = {}
for column in columns_to_process:
    nonzero_mean_values = df[df[column] != 0].groupby('Mã CK')[column].mean()
    symbol_nonzero_mean_values[column] = nonzero_mean_values

#Replace the zero values in each column with the corresponding average value of that code
for column in columns_to_process:
    for symbol, mean_value in symbol_nonzero_mean_values[column].items():
        mask = (df['Mã CK'] == symbol) & (df[column] == 0)
        df.loc[mask, column] = mean_value
        
df
```

```python
#Calculate the average values
mean_eps = df['EPS'].mean()
mean_bvps = df['BVPS'].mean()

#Create secondary DataFrame to store the conditions
conditions = pd.DataFrame({
    'ROE': df['ROE'] > 0.15,
    'ROA': df['ROA'] > 0.5,    
    'EPS': df['EPS'] > mean_eps,
    'DFL': (0 < df['DFL']) & (df['DFL'] < 1.1),
    'Current_Ratio': df['Current_Ratio'] > 1.5,
    'D/E': (1 < df['D/E']) & (df['D/E'] < 1.5),
    'P/E': (0 < df['P/E']) & (df['P/E'] < 9),
    'BVPS': df['BVPS'] > mean_bvps,
    'P/B': (0 < df['P/B']) & (df['P/B'] < 1.2)
})

df['Investment'] = (conditions.sum(axis=1) >=5).astype(int)

df
```

```python
investment_counts = df['Investment'].value_counts()
#Pie chart of Distribution of Investment
plt.figure(figsize=(6, 6))
plt.pie(investment_counts, labels=investment_counts.index, autopct='%1.2f%%', startangle=140)
plt.title('Distribution of Investment')
plt.axis('equal') 
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['Mã CK', 'Năm','Investment'], axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
y = df['Investment']
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, 
                                                    test_size = 0.3, 
                                                    random_state = 42)
```

# Logistic Regression

```python
#Train the Logistic Regression model on the training set
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)

#Predict labels for test data
y_pred = classifier.predict(X_test)

# Calculate the predicted probability
y_pred_proba = classifier.predict_proba(X_test)

#Print the prediction and the predicted probability
print('Predicted labels:')
print(y_pred[:10])
print('\nPredicted probabilities:')
print(y_pred_proba)

#Evaluate model performance
print('\nAccuracy Score:', accuracy_score(y_test, y_pred) * 100, '%')
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm_percentage = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

# Visualization Confusion Matrix of the Logistic Regression.
disp = ConfusionMatrixDisplay(confusion_matrix=cm_percentage, display_labels=classifier.classes_)
fig, ax = plt.subplots(figsize=(8, 8))
disp.plot(cmap=plt.cm.Blues, ax=ax, values_format='.2f')
plt.title('Confusion Matrix of Logistic Regression (Percentage)')
plt.show()
```

# KNN

```python
# Check k-value with accuracy
list_k = []
list_acc = []
for K_value in range (1, int(y_train.shape[0]**0.5)):
    list_k.append(K_value)
    classifier = KNeighborsClassifier(n_neighbors = K_value)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    acc = accuracy_score(y_test, y_pred)*100
    list_acc.append(acc)
    print('Accuracy is', accuracy_score(y_test, y_pred)*100, '% for k_value:', K_value)
```

```python
# Visualize k-value with accuracy
plt.figure(figsize = (12, 6))
plt.plot(list_k, list_acc)
plt.title('Accuracy vs K-value for KNN')
plt.xlabel('K-value')
plt.ylabel('Accuracy (%)')
plt.show()
```
